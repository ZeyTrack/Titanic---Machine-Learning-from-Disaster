## 2.   Feature Engineering


```python
def get_age_category(row):
  if row.Age <=12:
    return 'Children'
  elif row.Age <= 18:
    return 'Teenagers'
  elif row.Age <= 50:
    return 'Adults'
  else:
    return 'Older'

data['AgeCategory'] = data.apply(get_age_category, axis=1)
test['AgeCategory'] = test.apply(get_age_category, axis=1)

data["FamilySize"] = original_data.SibSp + original_data.Parch
test["FamilySize"] = test.SibSp + test.Parch

data.drop(['SibSp', 'Parch'], axis=1, inplace = True)
test.drop(['SibSp', 'Parch'], axis=1, inplace = True)

data['IsAlone'] = ['yes' if family_size == 0 else 'no' for family_size in data.FamilySize] #read on the comprehension list
test['IsAlone'] = ['yes' if family_size == 0 else 'no' for family_size in test.FamilySize]

data
```

##3.  Data Cleaning

```python
for col_name in test.columns:
  print(col_name)
  print(test[col_name].isna().sum())
```

```python
# you can drop before from the data before spliting it
data = data.drop(['PassengerId','Ticket','Cabin','Name'],axis = 1)
test = test.drop(['Ticket','Cabin','Name'],axis = 1)

data = data[data.Fare < 100]

# Split data into train and validation sets with shuffling
X_train, X_val, y_train, y_val = train_test_split(data.drop('Survived',axis=1), data['Survived'], test_size=0.15, shuffle=True, random_state=1) #random state for reproducing the same result every run

train = pd.concat([X_train,y_train],axis=1)
val = pd.concat([X_val,y_val],axis=1)

age_means = train.groupby('Survived')['Age'].mean()
X_train['Age'] =  train.apply(lambda row: age_means[row['Survived']] if pd.isnull(row['Age']) else row['Age'], axis=1)
X_val['Age'] =  val.apply(lambda row: age_means[row['Survived']] if pd.isnull(row['Age']) else row['Age'], axis=1)
test['Age'].fillna(X_train.Age.mean(), inplace = True)

test['Fare'].fillna(X_train.Fare.mean(), inplace = True)

X_train['Embarked'].fillna(X_train['Embarked'].mode()[0], inplace = True)
X_val['Embarked'].fillna(X_train['Embarked'].mode()[0], inplace = True)
```

## 4.   Data Preprocessing


```python
X_train = pd.get_dummies(X_train, columns = ['Sex','Embarked','IsAlone']) # one hot encoding
X_val = pd.get_dummies(X_val, columns = ['Sex','Embarked','IsAlone']) # one hot encoding
test = pd.get_dummies(test, columns = ['Sex','Embarked','IsAlone']) # one hot encoding

ordinal_encoder = OrdinalEncoder(categories = [['Children', 'Teenagers', 'Adults', 'Older']]) #adding categories order for the mapping
X_train['AgeCategory'] = ordinal_encoder.fit_transform(X_train[['AgeCategory']])
X_val['AgeCategory'] = ordinal_encoder.transform(X_val[['AgeCategory']])
test['AgeCategory'] = ordinal_encoder.transform(test[['AgeCategory']])

normalizer = MinMaxScaler()
for col_name in ['Age', 'Fare']:
  X_train[col_name] = normalizer.fit_transform(X_train[[col_name]])
  X_val[col_name] = normalizer.transform(X_val[[col_name]])
  test[col_name] = normalizer.transform(test[[col_name]])
```

```python
import xgboost as xgb
from sklearn.metrics import accuracy_score
```

```python
# Define the features and target variable
features = X_train.columns
target = 'Survived'

# Create DMatrix objects for XGBoost
dtrain = xgb.DMatrix(X_train[features], label=y_train)
dval = xgb.DMatrix(X_val[features], label=y_val)
dtest = xgb.DMatrix(test[features])

```

```python
# Define XGBoost parameters
params = {
    'objective': 'binary:logistic',
    'eval_metric': 'logloss',
    'max_depth': 3,
    'learning_rate': 0.1,
    'subsample': 0.8,
    'colsample_bytree': 0.8,
    'seed': 42
}

# Train the XGBoost model
num_round = 100
model = xgb.train(params, dtrain, num_round, evals=[(dval, 'validation')], early_stopping_rounds=10)
```

```python
# Predict on the test set
y_pred = model.predict(dtest)
# Convert probabilities to binary predictions
predictions = [1 if p >= 0.5 else 0 for p in y_pred]
```

```python
# Optionally, evaluate the model on the validation set
y_val_pred = model.predict(dval)
val_predictions = [1 if p >= 0.5 else 0 for p in y_val_pred]
accuracy = accuracy_score(y_val, val_predictions)
print(f'Validation Accuracy: {accuracy}')
```

## 4. *-Hyperparameter Tuning with RandomizedSearchCV:*






```python
from datetime import datetime
from sklearn.model_selection import RandomizedSearchCV, StratifiedKFold
from xgboost import XGBClassifier

def timer(start_time=None):
    if not start_time:
        start_time = datetime.now()
        return start_time
    elif start_time:
        thour, temp_sec = divmod((datetime.now() - start_time).total_seconds(), 3600)
        tmin, tsec = divmod(temp_sec, 60)
        print('\nTime taken: %i hours %i minutes and %s seconds.' % (thour, tmin, round(tsec, 2)))

params = {
    'min_child_weight': [1, 5, 10],
    'gamma': [0.5, 1, 1.5, 2, 5],
    'subsample': [0.6, 0.8, 1.0],
    'colsample_bytree': [0.6, 0.8, 1.0],
    'max_depth': [3, 4, 5]
}

xgb = XGBClassifier(learning_rate=0.02, n_estimators=600, objective='binary:logistic', silent=True, nthread=1)

folds = 3
param_comb = 5

skf = StratifiedKFold(n_splits=folds, shuffle=True, random_state=1001)

random_search = RandomizedSearchCV(xgb, param_distributions=params, n_iter=param_comb,
                                   scoring='roc_auc', n_jobs=4, cv=skf.split(X_train_scaled, y_train),
                                   verbose=3, random_state=1001)

start_time = timer(None)
random_search.fit(X_train_scaled, y_train)
timer(start_time)

print('\nAll results:')
print(random_search.cv_results_)
print('\nBest estimator:')
print(random_search.best_estimator_)
print('\nBest normalized gini score for %d-fold search with %d parameter combinations:' % (folds, param_comb))
print(random_search.best_score_ * 2 - 1)
print('\nBest hyperparameters:')
print(random_search.best_params_)
```
